```python
import numpy as np
import pandas as pd
import sklearn
```

```python
from sklearn.datasets import load_diabetes
df = load_diabetes()
```

```python
df.keys()
```

```python
print(df.data) 
```

```python
diabetes = pd.DataFrame(df.data, columns = df.feature_names)
diabetes.head()
```

```python
diabetes['TARGET'] = df.target
diabetes.head()
```

```python
diabetes.isnull()
```

```python
diabetes.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
q = diabetes.drop('TARGET', axis = 1)
r = diabetes['TARGET']

q_train, q_test, r_train, r_test = train_test_split(q, r, test_size = 0.15, random_state = 5)

print(q_train.shape)
print(q_test.shape)
print(r_train.shape)
print(r_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lin_model = LinearRegression()

lin_model.fit(q_train, r_train)
```

```python
R_train_predict = lin_model.predict(q_train)
rmse = (np.sqrt(mean_squared_error(r_train, R_train_predict)))

print("The model performance for training set")
print('RMSE is {}'.format(rmse))
print("\n")



#on testing set
R_test_predict = lin_model.predict(q_test)
rmse = (np.sqrt(mean_squared_error(r_test, R_test_predict)))


print("The model performance for testing set")
print('RMSE is {}'.format(rmse))
```
